```python
'''Take the choice of subject from the students having grade FF. the choice can
be NLP/NNDL/TC/Big data and alott accordingly
the choice should be asked like
Abhay, roll no BT20CSE012 what subject do you want  NLP/NNDL/TC/Big data?'''
```

```python
df2 = df[(df['Grades'] == 'FF')]
```

```python
df2.head()
```

```python
ffname=df2['Name']
ffroll=df2['Roll']
print(type(ffname))
```

```python
ffname=list(ffname)
ffroll=list(ffroll)
for i in range(0,len(ffroll)):
    print(ffname[i],' roll no',ffroll[i],'what subject do you want  NLP/NNDL/TC/Big data?')
    print('1 for NLP,2 for NNDL, 3 for TC, 4 for Big data')
    choice=int(input())
    if choice==1:
        df.loc[df['Roll']==ffroll[i],'Subject']='NLP'
    elif choice==2:
        df.loc[df['Roll']==ffroll[i],'Subject']='NNDL'
    elif choice==3:
        df.loc[df['Roll']==ffroll[i],'Subject']='TC'
    elif choice==4:
        df.loc[df['Roll']==ffroll[i],'Subject']='Big data'
    else:
        print('Wrong choice')
```
