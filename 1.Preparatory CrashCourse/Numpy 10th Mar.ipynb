```python
#How many total products are sold by the company
len(number_array)
```

```python
# how many items were sold in total
np.sum(number_array)
```

```python
# what is the average price of the product sold by the company
np.mean(price_array)
```

```python
# what is the price of the costliest item sold
np.max(price_array)
```

```python
# what is the total cost of the company [revenu = price * quantity]
np.sum(number_array * price_array)
```

```python
# can you confirm demand for 2oth product is more than 50th product [True/false]
demand_20th_product = number_array[19]  # Index 19 corresponds to the 20th product
demand_50th_product = number_array[49]  # Index 49 corresponds to the 50th product

demand_20th_greater_than_50th = demand_20th_product > demand_50th_product
print("Demand for 20th product is greater than 50th product:", demand_20th_greater_than_50th)
```

```python
# how many product fall under category of expensive goods
# an Expensive good is the price is more than the average price of product sold by the company
average_price = np.mean(price_array)

expensive_goods_count = np.sum(price_array > average_price)

print("Number of products considered expensive:", expensive_goods_count)
```
