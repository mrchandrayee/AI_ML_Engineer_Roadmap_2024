```python
# marks as an input from user : marks > 80 : Print('Excellent '), marks 60 to 80
# marks 40 to 70: ('Fair') else print('can do better')


marks = float(input("Enter the marks: "))

if marks > 80:
    print('Excellent')
elif 60 <= marks <= 80:
    print('Good')
elif 40 <= marks < 60:
    print('Fair')
else:
    print('Can do better')
```

```python
# program to check whether the given integer is multiple of 5 and 7

# Get an integer as input from the user
num = int(input("Enter an integer: "))

# Check if the number is divisible by both 5 and 7
if num % 5 == 0 and num % 7 == 0:
    # Print that the number is multiple of 5 and 7
    print("Number is multiple of 5 and 7")
else:
    # Print that the number is not multiple of 5 and 7
    print("Number is not multiple of 5 and 7")
```

```python
# python program to display all the multiples of 3 in a range 10 to 50

# Define the range of numbers
start = 10
end = 50

# Loop through the range
for num in range(start, end + 1, 5):
    # Check if the number is divisible by 3
    if num % 3 == 0:
        # Print the number
        print(num)
```

```python
# Get an integer as input from the user and store it in a variable called n
n = int(input("Enter an integer: "))
sum_result = 0
for i in range(1, n + 1):

    sum_result += i

print("The sum of integers from 1 to", n, "is:", sum_result)
```

```python
# python code to calculate the sum of all number from 1 to a given number by user

# Get an integer as input from the user
n = int(input("Enter a positive integer: "))

# Initialize the sum variable
sum = 0

# Loop from 1 to n
for i in range(1, n + 1):
    # Add i to the sum
    sum = sum + i

# Print the sum
print("The sum of all numbers from 1 to", n, "is", sum)
```

```python
# Given a dictionary of employee and tenure
# We need to print those employees whose tenure > 5

emp_list = [
    {'emp_name': 'Alice', 'tenure': 7},
    {'emp_name': 'Bob', 'tenure': 2},
    {'emp_name': 'Charlie', 'tenure': 13},
    {'emp_name': 'David', 'tenure': 7},
    {'emp_name': 'John', 'tenure': 9}
]

# Print employees whose tenure > 5
for emp in emp_list:
    if emp['tenure'] > 5:
        print(f"{emp['emp_name']} has tenure greater than 5 years.")
```

```python
# given a dictionary of employee and tenure,
# we need to print those empl whose tenure > 5

# Define the dictionary of employee and tenure
emp = [
    {'emp_name': 'Alice', 'tenure': 7},
    {'emp_name': 'Bob', 'tenure': 2},
    {'emp_name': 'Charlie', 'tenure': 13},
    {'emp_name': 'David', 'tenure': 7},
    {'emp_name': 'John', 'tenure': 4}
]

# Loop through the dictionary
for e in emp:
    # Get the name and tenure of each employee
    name = e['emp_name']
    tenure = e['tenure']
    # Check if the tenure is greater than 5
    if tenure > 5:
        # Print the name of the employee
        print(name)
```
